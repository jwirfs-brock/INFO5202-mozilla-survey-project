```python
import pandas as pd
import numpy as np
```

```python
# Import Mozilla survey data
pd.options.display.max_columns = 999
pd.options.display.max_rows= 99999
df = pd.read_csv('mozilla-survey.csv', encoding = "ISO-8859-1", low_memory= False)
```

```python
df.head()
```

```python
# Drop rows with blanks in the country column
df.dropna(subset=['Country'], inplace=True)
```

```python
df.describe()
```

```python
# Count how many responses came from each country
df0 = df.groupby(['Country'], as_index=False).count()
```

```python
# Make a data frame that is just country and survey count
survey_count = df0[['Country','Response ID']]
survey_count = survey_count.rename(columns={"Response ID":"Count"})
# Sort by count
survey_count.sort_values(by=['Count'])
```

```python
# Rename columns about which terms you can explain
df1 = df.rename(columns={"IoT:Check all the terms below that you could explain to a friend:":"IoT","Connected Devices:Check all the terms below that you could explain to a friend:" : "Connected Devices","Botnet:Check all the terms below that you could explain to a friend:":"Botnet","Blockchain:Check all the terms below that you could explain to a friend:":"Blockchain","RFID:Check all the terms below that you could explain to a friend:":"RFID","DDOS:Check all the terms below that you could explain to a friend:":"DDOS","Zero Day:Check all the terms below that you could explain to a friend:":"Zero Day","VPN:Check all the terms below that you could explain to a friend:":"VPN","TOR:Check all the terms below that you could explain to a friend:":"TOR","I donÛªt know what any of these things are:Check all the terms below that you could explain to a friend:":"None"})
```

```python
# Select just the basic info + what terms you can explain and create a new data frame
df2 = df1[['Response ID','Time Started','Date Submitted','Status','Language','Country','Link Name','I consider myself:','IoT','Connected Devices','Botnet','Blockchain','RFID','DDOS','Zero Day','VPN','TOR','None']]
```

```python
# Replace names of terms in "what can you explain" columns with 0s and 1s
# Do this so we can calculate a frequency of how many respondents can explain each term
df2.replace(to_replace="IoT", value=1, inplace=True)
df2.replace(to_replace="Connected Devices", value=1, inplace=True)
df2.replace(to_replace="Botnet", value=1, inplace=True)
df2.replace(to_replace="Blockchain", value=1, inplace=True)
df2.replace(to_replace="RFID", value=1, inplace=True)
df2.replace(to_replace="DDOS", value=1, inplace=True)
df2.replace(to_replace="Zero Day", value=1, inplace=True)
df2.replace(to_replace="VPN", value=1, inplace=True)
df2.replace(to_replace="TOR", value=1, inplace=True)
df2.replace(to_replace="I don\x89Ûªt know what any of these things are", value=1, inplace=True)
```

```python
# Fill NaN values with 0s
df2.fillna(value=0, inplace=True)
```

```python
df2.head()
```

```python
# When looking through the data, I noticed that there are some survey respondents who selected
# "I don't know what any of these things are" as well as other terms.
# In the case there they only selected a subset, I took these as valids.
# But in many cases, they checked ALL the boxes -- so I am making the assumption that this data is invalid.
# Drop rows where ALL values are checked, including "I don't know what any of these things are"
# by looking for rows where the sume of all of these items is 10.
# Create a new column that tells you how many terms each survey respondent could explain

# Create a column that is summing all the check boxes, INCLUDING "I don't know"
df2['Total_Checked'] = df2['IoT'] + df2['Connected Devices'] + df2['Botnet'] + df2['Blockchain'] + df2['RFID'] + df2['DDOS']+ df2['Zero Day'] + df2['VPN']+ df2['TOR']+ df2["None"]
# Check for values that are 10, and drop those
```

```python
df2 = df2[df2.Total_Checked != 10]
```

```python
# Create a new column that tells you how many terms each survey respondent could explain
df2['Number of Terms'] = df2['IoT'] + df2['Connected Devices'] + df2['Botnet'] + df2['Blockchain'] + df2['RFID'] + df2['DDOS']+ df2['Zero Day'] + df2['VPN']+ df2['TOR']
```

```python
# Groupby country based on the mean (which gives the average for each term, and average number of terms)
df3 = df2.groupby(['Country'], as_index=False).mean()
# Drop Response ID column (because it's meaningless once we group using the mean)
df3.drop(['Response ID'], axis=1)
```

```python
df3
```

```python
# Find the term people are most likely to be able to explain in each country
df3['Most_Common_Term'] = df3[['IoT','Connected Devices','Botnet','Blockchain','RFID','DDOS','Zero Day','VPN','TOR']].idxmax(axis=1)
# And add a column that has the frequency of that term
df3['Most_Common_Value'] = df3[['IoT','Connected Devices','Botnet','Blockchain','RFID','DDOS','Zero Day','VPN','TOR']].max(axis=1)
df3['MCV_formatted'] = df3['Most_Common_Value']*100

# Find the term people are most likely to be able to explain in each country
df3['Least_Common_Term'] = df3[['IoT','Connected Devices','Botnet','Blockchain','RFID','DDOS','Zero Day','VPN','TOR']].idxmin(axis=1)
df3['Least_Common_Value'] = df3[['IoT','Connected Devices','Botnet','Blockchain','RFID','DDOS','Zero Day','VPN','TOR']].min(axis=1)
df3['LCV_formatted'] = df3['Least_Common_Value']*100
```

```python
df4 = df3.join(survey_count.set_index('Country'), on='Country', how='left', sort=False, lsuffix='', rsuffix='1')
```
